# <font color='Orange'>Requirements specified for Web Scraping</font>


Import all libraries and modules that you are going to use.

```python
import numpy as np                                   # za manipulacija so vrednosti i listi
import pandas as pd                                  # za manipulacija so DataFrames
import requests                                      # za prakjanje i analiza na HTTP requests 
from bs4 import BeautifulSoup                        # za simnuvanje HTML i XML podatoci od strani
from IPython.display import HTML                     # za ubav prikaz na HTML podatoci
import warnings
import math

requests.packages.urllib3.disable_warnings()
warnings.filterwarnings("ignore")
```

# <font color = 'Orange'> Start the scraping by sending HTTP request</font> 

 **Fetch a website's content - send request**

```python
url = 'http://diplomski.finki.ukim.mk/StartPage/DiplomaList'
response = requests.get(url)
```

**Get the response-status code and see:**
<li>If the code starts with a 2, it means the request was successful.



```python
response
```

# <font color = 'orange'>Parse the HTML with BeautifulSoup</font>

Beautiful Soup is a Python library for pulling data out of HTML and XML files. It works with your favorite parser to provide idiomatic ways of navigating, searching, and modifying the parse tree. It commonly saves programmers hours or days of work. 


**Parse using BeautifulSoap**




```python
rawHTML = response.text
cleanHTML = BeautifulSoup(rawHTML, 'html.parser')
```

**Print the HTML code that you get**

```python
cleanHTML
```

# <font color='Orange'>Extracting data</font>

From the printed HTML code search for the data you need for building the dataset, find the classes for groupby option.

**In the following text-box write the attributes you are going to use**

<font color='red'> You need to extract a minimum of four features (lists/columns) from the web site (example: titles, posting time, comments, likes, ...)</font>

Атрибути кои ќе се вадат се: Наслов на труд, име на студент, индекс, ментор, член 1, член 2, датум на одбрана и време на одбрана.

```python
# Vadenje naslovi na trudovite
thesisTitles = []
thesisTitleBlocks = cleanHTML.select('.panel-heading')
#thisisTitleBlocks[0].select_one('b').text
for block in thesisTitleBlocks:
    thesisTitles.append( block.select_one('b').text )
thesisTitles[0:5]
```

```python
# Vadenje na ostatokot info. od tabelite za diplomski trudovi
thesisInfo = []             # lista koja kje sodrzi info. za poseben naucen trud
tablesInfo = []             # lista od listite za naucnite trudovi

for table in cleanHTML.find_all("table"):
    for i in table.find_all("td", style="width: 30%"):
        thesisInfo.append( str( i.find_next_sibling("td").select_one("strong").text ) )
    tablesInfo.append( thesisInfo )
    thesisInfo = []

thesisInfo.clear()
tablesInfo[0:3]
```

**Print all lists that are containing the elements**

```python
# Listi so vrednosti za atributite
studentNames = []
studentIDs = []
mentors = []
members = []
defendingDates = []
defendingTimes = []

for table in tablesInfo:
    studentInfo = table[0].split(" - ")
    studentIDs.append( studentInfo[0] )
    studentNames.append( studentInfo[1] )
    
    mentors.append( table[1] )
    
    members.append( table[2] + " , " + table[3] )
    
    if( table[4]=="" or table[4]==" " ):
        table[4] = math.nan
    defendingDates.append( table[4] )
    
    if( table[5]=="" or table[5]==" " ):
        table[5] = math.nan
    defendingTimes.append( table[5] )
```

**Check if all lists are the same size**

```python
len(thesisTitles) == len(studentNames) == len(studentIDs) == len(mentors) == len(members) == len(defendingDates) == len(defendingTimes)
```

# <font color='Orange'>Create a Pandas DataFrame</font> 


**Create the DataFrame from the lists you get and name your *Columns***

```python
thesisDict = {
    "Наслов": thesisTitles,
    "Студент": studentNames,
    "Индекс": studentIDs,
    "Ментор": mentors,
    "Членови": members,
    "Датум на одбрана": defendingDates,
    "Време на одбрана": defendingTimes
}
df = pd.DataFrame(thesisDict)
```

**Print the dataset you obtain**

```python
df.to_csv('')
df
```
